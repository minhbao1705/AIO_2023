```python
import random

# predict function
def predicted_func(x, a, b):
  return a*x + b

# parameters
num_steps = 100
lr = 0.01

# give data
x1 = -2
y1 = -1

x2 = 5
y2 = 13

# 1. set a, b randomly
a = random.random()*10.0 - 5.0
b = random.random()*10.0 - 5.0
print(a, b)

for i in range(num_steps):
  # -------- for the first point ----------
  x_value, y_value = x1, y1

  # compute predicted_y
  predicted_y = predicted_func(x_value, a, b)

  # compute g
  g_value = (predicted_y - y_value)**2

  # compute partial gradients for a and b
  dg_da = 2*x_value*(predicted_y - y_value)
  dg_db = 2*(predicted_y - y_value)

  # update
  a = a - lr*dg_da
  b = b - lr*dg_db

  # print
  if i%10==0:
    print(f'a: {a} and b: {b}')
    print(f'g_value 1: {g_value}')
    print('--------------')

  # -------- for the second point ----------
  x_value, y_value = x2, y2
  # compute predicted_y
  predicted_y = predicted_func(x_value, a, b)

  # compute g
  g_value = (predicted_y - y_value)**2

  # compute partial gradients for a and b
  dg_da = 2*x_value*(predicted_y - y_value)
  dg_db = 2*(predicted_y - y_value)

  # update
  a = a - lr*dg_da
  b = b - lr*dg_db

  # print
  if i%10==0:
    print(f'a: {a} and b: {b}')
    print(f'g_value 2: {g_value}')
    print('--------------')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# f function
def blackbox_func(x):
  return 2*x - 3

def func1(x):
  return 2.008*x + 2.947

# plot f_func
x_list = np.arange(-10, 10, 0.1).tolist()
fx_list = [blackbox_func(x) for x in x_list]
fx1_list = [func1(x) for x in x_list]

plt.plot(x_list, fx_list, 'r')
plt.plot(x_list, fx1_list, 'g')
plt.show()
```
