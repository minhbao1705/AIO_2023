```python
import numpy as np

data = np.array([5, 3, 6, 7, 4])
mean = data.mean()
var = data.var()
print(mean)
print(var)
```

```python
import math

def gaussian_function(data, mean, var):
    return (1.0/np.sqrt(2*math.pi)*var) * np.exp(-(data-mean)**2 / (2*var))
```

```python
gaussian_function(data, mean, var)
```

```python
import numpy as np

data = np.array([1.4, 1.0, 1.3, 1.9, 2.0])
mean = data.mean()
var = data.var()
print(mean)
print(var)
```

```python
import math

def gaussian_function(data, mean, var):
    return (1.0/np.sqrt(2*math.pi)*var) * np.exp(-(data-mean)**2 / (2*var))
```

```python
import numpy as np

data_c1 = np.array([1.4, 1.0, 1.3, 1.9, 2.0])
mean_c1= data_c1.mean()
var_c1 = data_c1.var()
print(mean_c1)
print(var_c1)
print(gaussian_function(data_c1, mean_c1, var_c1))
```

```python
import numpy as np

data_c2 = np.array([3.8, 4.1, 3.9, 4.2, 3.4])
mean_c2= data_c2.mean()
var_c2 = data_c2.var()
print(mean_c2)
print(var_c2)
print(gaussian_function(data_c2, mean_c2, var_c2))
```

```python
x_data = np.arange(0.1, 5, 0.01)
y0_data = gaussian_function(x_data, mean_c1, var_c1)
y1_data = gaussian_function(x_data, mean_c2, var_c2)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_data, y0_data)
plt.plot(x_data, y1_data)
```
