# Compute the circumference of a unit circle

```python
import math

def compute_length(selected_degree):
  # sin value for the selected_degree
  # <--> the length of a part
  sin = math.sin(math.radians(selected_degree))

  # summarize (360/selected_degree) parts
  length = sin * (360/selected_degree)

  return length
```

```python
print(compute_length(10)/2)
print(compute_length(1)/2)
print(compute_length(0.1)/2)
```

```python
print(compute_length(0.01)/2)
```

# Compute the area of a unit circle (1)

```python
import math

def compute_area(selected_degree):
  # sin and cosine values for the selected_degree
  sin_degree = math.sin(math.radians(selected_degree))
  cos_degree = math.cos(math.radians(selected_degree))

  # compute area for a part
  area_degree = sin_degree*cos_degree/2

  # summarize (360/selected_degree) parts
  area = area_degree * (360/selected_degree)
  return area
```

```python
print(compute_area(10))
print(compute_area(1))
print(compute_area(0.1))
print(compute_area(0.01))
```

# Compute the area of a unit circle (2)

```python
# The above part of a unit circle
import matplotlib.pyplot as plt
import math
import numpy as np

def compute_y(x):
  return math.sqrt(1 - x*x)

data_x = np.arange(-1, 1, 1e-5).tolist()
data_y = [compute_y(x) for x in data_x]

plt.plot(data_x, data_y)
plt.show()
```

```python
# The below part of a unit circle
import matplotlib.pyplot as plt
import math

def compute_y(x):
  return -math.sqrt(1 - x*x)

data_x = np.arange(-1, 1, 1e-5).tolist()
data_y = [compute_y(x) for x in data_x]
plt.plot(data_x, data_y)
plt.show()
```

```python
import math
import numpy as np

def get_y(x):
  y = math.sqrt(1 - x*x)
  return y

# create x, y
step = 1e-5
x_data = np.arange(-1, 1, step).tolist()
y_data = [get_y(x) for x in x_data]

# compute area
areas = [y*step for y in y_data]
area = sum(areas)
print(area*2)
```

# Compute the are of a unit cricle using its radius

```python
import math
import numpy as np

step = 1e-5
radii = np.arange(0, 1, step).tolist()

areas = [math.pi*2*radius*step for radius in radii]
area = sum(areas)
print(f'area is {area}')
```
