```python
import matplotlib.pyplot as plt
import math
import numpy as np

def plot_function_and_derivative(x, f_x, df_x):
  # setting the axes at the centre
  fig, ax = plt.subplots(figsize=(10, 5))
  ax.spines['left'].set_position('center')
  ax.spines['bottom'].set_position('zero')
  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')
  ax.xaxis.set_ticks_position('bottom')
  ax.yaxis.set_ticks_position('left')

  # plot the function
  ax.plot(x, f_x, color="#d35400", linewidth=3, label='Function')
  ax.plot(x, df_x, color="#1abd15", linewidth=3, label='Derivative')

  ax.legend(loc="upper left", frameon=False)
  plt.show()
```

### using derivative formula

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

def d_sigmoid(x):
  result = sigmoid(x)*(1-sigmoid(x))
  return result
```

```python
import numpy as np

x_data = np.arange(-10, 10, 0.1).tolist()
y_data = [sigmoid(x) for x in x_data]
dy_data = [d_sigmoid(x) for x in x_data]

plot_function_and_derivative(x_data, y_data, dy_data)
```

### using autograd

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

def autograd(x, epsilon=1e-6):
  delta_y = sigmoid(x+epsilon) - sigmoid(x)
  return delta_y/epsilon
```

```python
x_data = np.arange(-10, 10, 0.1).tolist()
y_data = [sigmoid(x) for x in x_data]
dy_data = [autograd(x) for x in x_data]

plot_function_and_derivative(x_data, y_data, dy_data)
```

### autograd using numpy

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
import numpy as np

x_data = np.arange(-10, 10, 0.1).tolist()
y_data = [sigmoid(x) for x in x_data]
dy_data = np.gradient(y_data, x_data).tolist()

plot_function_and_derivative(x_data, y_data, dy_data)
```
